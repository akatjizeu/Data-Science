```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('Train_dataset.csv')
```

```python
test = pd.read_csv('Test_dataset.csv')
df_test = test.copy()
```

```python
# Preprocessing (Handle missing values)
#df.fillna(df.median(), inplace=True)  # Fill numerical columns with median values

# Preprocessing (Handle missing values)
# Select only numerical columns before calculating median
numerical_df = df.select_dtypes(include=np.number)
#numerical_df_test = df_test.select_dtypes(include=np.number)

# Fill numerical columns with median values
df[numerical_df.columns] = df[numerical_df.columns].fillna(numerical_df.median())
#df_test[numerical_df_test.columns] = df_test[numerical_df_test.columns].fillna(numerical_df_test.median())
```

```python
# One-Hot Encoding for categorical variables
categorical_cols = ['Cuisine', 'City', 'Restaurant Location', 'Endorsed By']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

#df_test = pd.get_dummies(df_test, columns=categorical_cols, drop_first=True)
```

```python

# USe the train data but do not split for training
X_train = df.drop('Annual Turnover', axis=1)
y_train = df['Annual Turnover']

#use test data fo testing
#X_test = df_test.copy()
#y_test = df_test['Instagram Popularity Quotient']
```

```python
# Feature scaling (optional but can improve performance for some models)
#scaler = StandardScaler()
#X_train = scaler.fit_transform(X_train)
#X_test = scaler.transform(X_test)

# Feature scaling (optional but can improve performance for some models)
scaler = StandardScaler()

# Select only numerical features for scaling
numerical_features = X_train.select_dtypes(include=np.number).columns
X_train_numerical = X_train[numerical_features]
#X_test_numerical = X_test[numerical_features]


X_train[numerical_features] = scaler.fit_transform(X_train_numerical)
#_test[numerical_features] = scaler.transform(X_test_numerical)
```

```python
# Model: Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_numerical, y_train)
```

```python
# Predict and evaluate
y_pred_train = model.predict(X_train_numerical)
#y_pred_test = model.predict(X_test_numerical)

#df_test['Annual Turnover'] = y_pred_test

# Calculate RMSE
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
#rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

print(f'Training RMSE: {rmse_train}')
#print(f'Testing RMSE: {rmse_test}')
```

```python
#prediction = model.predict(df_test[['Instagram Popularity Quotient']])
#prediction

# Select the numerical features that the model was trained on
numerical_features = X_train.select_dtypes(include=np.number).columns

# Filter the test data to include only these features
X_test_numerical = df_test[numerical_features]

# Predict using the filtered test data
prediction = model.predict(X_test_numerical)
prediction
```
