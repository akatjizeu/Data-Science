```python
# Assuming 'train' is your training dataset and 'test' is your test dataset
# Step 1: Check for missing values
missing_values = train.isnull().sum()

# Step 2: Convert date field into useful numerical features
train['Opening Day of Restaurant'] = pd.to_datetime(train['Opening Day of Restaurant'], format='%d-%m-%Y')
train['Years Since Opening'] = 2024 - train['Opening Day of Restaurant'].dt.year

# Step 3: Encode categorical variables
categorical_cols = ['Cuisine', 'City', 'Restaurant Location', 'Endorsed By']

# ----> Modified section starts here <----
# Create a function to split comma-separated values into multiple columns
def split_and_encode(df, column):
    # Split the comma-separated values in the column
    split_values = df[column].str.split(',', expand=True)

    # Create new column names for each split value
    new_cols = [f"{column}_{value}" for value in split_values.stack().unique()]

    # One-hot encode the split values
    encoded_values = pd.get_dummies(split_values.stack()).groupby(level=0).sum()  # Sum to handle duplicates
    encoded_values.columns = new_cols

    # Concatenate the encoded values with the original DataFrame
    df = pd.concat([df, encoded_values], axis=1)

    # Drop the original column
    df = df.drop(columns=[column])

    return df

# Apply the function to the 'Cuisine' column
train_encoded = split_and_encode(train, 'Cuisine')

# One-hot encode the other categorical columns
train_encoded = pd.get_dummies(train_encoded, columns=[col for col in categorical_cols if col != 'Cuisine'], drop_first=True)
# ----> Modified section ends here <----

# Drop unnecessary columns like 'Registration Number' and 'Opening Day of Restaurant' to avoid irrelevant features
train_encoded = train_encoded.drop(['Registration Number', 'Opening Day of Restaurant'], axis=1)

# Step 4: Fill missing values with median for numerical features
# Get a list of numerical columns
numerical_cols = train_encoded.select_dtypes(include=['number']).columns

# Fill missing values only in numerical columns with their median
train_encoded[numerical_cols] = train_encoded[numerical_cols].fillna(train_encoded[numerical_cols].median())

# Step 5: Scale numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(train_encoded[numerical_cols]) # Scale only numerical features

# Create a trainFrame with scaled numerical features
scaled_train = pd.DataFrame(scaled_features, columns=numerical_cols, index=train_encoded.index)

# Concatenate scaled numerical features with the remaining columns of data_encoded
train_processed = pd.concat([train_encoded.drop(columns=numerical_cols), scaled_train], axis=1)

#
```
